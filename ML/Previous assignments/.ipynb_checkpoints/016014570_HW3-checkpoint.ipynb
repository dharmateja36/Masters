e. There are a lot of zero entries in the spam bag-of-words dataset. This is indicative of sparse data and is typical for text analytics. Cosine similarity is often a good choice when dealing with sparse non-binary data. What target level would a 3-NN model using cosine similarity return for the query?

To calculate the cosine similarity between a query and its instances, one must calculate the vector lengths of each instance. In the next step, we will calculate the dot product between the query and the instances. In a final step, we divide the dot products by their respective vector lengths to calculate the cosine similarity.


```python
a= np.array((0,1,1,0,0,1,1))

b1 = np.array((3,0,0,0,0,0,0))
id1numerator=((0*3)+(0*1)+(0*1)+(0*0)+(0*0)+(0*1)+(0*1))
id1denominator=((np.sqrt(np.square(0) + np.square(1) + np.square(1) + np.square(0) + np.square(0) + np.square(1)+ np.square(1)))*
(np.sqrt(np.square(3) + np.square(0) + np.square(0) + np.square(0) + np.square(0) + np.square(0) + np.square(0))))
id1cosinesum=id1numerator/id1denominator
print(f'ID1 Cosine similarity ={id1cosinesum:.4f}')

b2 = np.array((1,2,1,1,1,0,0))
id2numerator=((1*0)+(2*1)+(1*1)+(1*0)+(1*0)+(0*1)+(0*1))
id2denominator=((np.sqrt(np.square(0) + np.square(1) + np.square(1) + np.square(0) + np.square(0) + np.square(1)+ np.square(1)))*
(np.sqrt(np.square(1) + np.square(2) + np.square(1) + np.square(1) + np.square(1) + np.square(0) + np.square(0))))
id2cosinesum=id2numerator/id2denominator
print(f'ID2 Cosine similarity ={id2cosinesum:.4f}')

b3 = np.array((0,0,1,1,1,0,0))
id3numerator=((0*0)+(0*1)+(1*1)+(1*0)+(1*0)+(0*1)+(0*1))
id3denominator=((np.sqrt(np.square(0) + np.square(1) + np.square(1) + np.square(0) + np.square(0) + np.square(1)+ np.square(1)))*
(np.sqrt(np.square(0) + np.square(0) + np.square(1) + np.square(1) + np.square(1) + np.square(0) + np.square(0))))
id3cosinesim=id3numerator/id3denominator
print(f'ID3 Cosine Similarity ={id3cosinesim:.4f}')

b4 = np.array((0,0,1,0,3,1,1))
id4numerator=((0*0)+(0*1)+(1*1)+(0*0)+(3*0)+(1*1)+(1*1))
id4denominator=((np.sqrt(np.square(0) + np.square(1) + np.square(1) + np.square(0) + np.square(0) + np.square(1)+ np.square(1)))*
(np.sqrt(np.square(0) + np.square(0) + np.square(1) + np.square(0) + np.square(3) + np.square(1) + np.square(1))))
id4cosinesim=id4numerator/id4denominator
print(f'ID4 Cosine Similarity ={id4cosinesim:.4f}')

b5 = np.array((0,1,0,0,0,1,1))
id5numerator=((0*0)+(1*1)+(0*1)+(0*0)+(0*0)+(1*1)+(1*1))
id5denominator=((np.sqrt(np.square(0) + np.square(1) + np.square(1) + np.square(0) + np.square(0) + np.square(1)+ np.square(1)))*
(np.sqrt(np.square(0) + np.square(1) + np.square(0) + np.square(0) + np.square(0) + np.square(1) + np.square(1))))
id5cosinesim=id5numerator/id5denominator
print(f'ID5 Cosine Similarity ={id5cosinesim:.4f}')
```

By comparing the cosine similarity values, we can conclude that the ID_5, ID_2, and ID_4 instances are the closest neighbors. There are three spam values: false, true, and false, with false being the majority. Therefore, the query will return false.

Could you implement your own KNN function from scratch (without 
calling any existing classifier packages)? Your own KNN function must have following parameters 
to tune: n_neighbors, weights. Can you compare your own classifier with sklearn 
neighbors.KNeighborsClassifier in terms performance, complexity, etc. Data usage: 
datasets.load_iris() from sklearn. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
from scipy.stats import mode
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
def distance(d1, d2):
    d1 = np.array(d1) 
    d2 = np.array(d2)
    dist=np.linalg.norm(d1 - d2)
    
    return dist
def Neighbours(data,k):
        distances=[]
        for index in range(len(x_train)):
            d=distance(x_train.iloc[index],data)
            distances.append((d,y_train.iloc[index]))
        distances.sort(key=lambda x:x[0])
        return distances[:k]
y_predict=[]
for i in range(len(x_test)):
    data=x_test.iloc[i]
    data=np.array(data)
    nei=Neighbours(data,5)
    y_predict.append( mode( nei )[0][0])
for i in range(len(y_predict)):
    y_predict[i]=y_predict[i][1]
    
y_prediction=np.array(y_predict).reshape(1,45)
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(y_test, y_predict)
```

Weighted KNN

```python
import numpy as np
from scipy.stats import mode
from numpy.random import randint
def distance(d1, d2):
    d1 = np.array(d1) 
    d2 = np.array(d2)
    dist=np.linalg.norm(d1 - d2)
    return dist 
def Join(a, b):
    joined = [(a[i], b[i]) for i in range(0, len(a))]
    return joined
def predict(x_train, y , x_test, k, w):
    label = []
    for i in x_test: 
        dist_p = []
        for j in range(len(x_train)): 
            distances = distance(np.array(x_train[j,:]) , i)
            dist_p.append(distances) 
        dist_p = np.array(dist_p) 
        dist = np.argsort(dist_p)[:k] 
        labels = y[dist]
        l = mode(labels) 
        l = l.mode[0]
        label.append(l)
        joined=Join(dist, labels)
        setosa = 0
        Versicolor = 0
        Verginica = 0
        for d in joined: 
            if d[1] == 0: 
                setosa += (1 / d[0]) 

            elif d[1] == 1: 
                Versicolor += (1 /d[0]) 
            else:
                Verginica +=  (1 /d[0])
        if (Versicolor > setosa) and (Versicolor > Verginica):
               n = 'versicolor'
        elif (setosa > Versicolor) and (setosa > Verginica):
               n = 'setosa'
        else:
               n = 'virginica'
            
    return label
a= iris.data
b= iris.target
xtrain,xtest,ytrain,ytest=train_test_split(a,b,test_size=0.3,random_state=0)
y_pred = predict(xtrain,ytrain,xtest ,10, 'distance')
accuracy_score(y_test, y_pred)
```
