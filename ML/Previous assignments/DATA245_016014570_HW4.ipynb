#### MultinomialNB
Mapped to 0, 1 & 2

```python

from sklearn.naive_bayes import MultinomialNB
import pandas as pd
X_mapped_train =  pd.DataFrame(mini_train_data)
X_mapped_test =  pd.DataFrame(dev_data)

def normalizer(val):
    if val >= 0.66:
      return 2.0
    elif val >= 0.33:
      return 1.0
    elif val < 0.33 and val >= 0:
      return 0.0
    else:
      return val
    
vn = np.vectorize(normalizer)

for column in X_mapped_train.columns.to_list():
  column_data_train = X_mapped_train[column].to_numpy()
  column_data_test = X_mapped_test[column].to_numpy()
  result_train = vn(column_data_train)
  result_test = vn(column_data_test)
  X_mapped_train[column].update(pd.Series(result_train))
  X_mapped_test[column].update(pd.Series(result_test))
```

Multinomial version do not increase the accuracy because <font color="Grey">the dataset is not of a discrete value type</font>, the method applied to map the values among 0, 1 & 2 will effect the accuracy as <font color="Grey"><b>we are introducing discreteness</b></font> in the dataset.
</font>
