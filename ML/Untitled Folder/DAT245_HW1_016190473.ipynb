```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame([42,47,59,27,84,49,72,43,73,59,58,82,50,79,89,75,70,59,67,35],columns=['score'])
high=1
low=0
normal_values = []

for i in df['score']:
    normalized_scores = ((i - min(df['score']))/((max(df['score'])-min(df['score'])))*(high - low)) + low
    normal_values.append(normalized_scores)

n_df=pd.DataFrame(normal_values,columns=['score'])
n_df
```

9a) Missing Values

Missing values are a common issue in ABT and can be considered as noise in the data. In the provided data quality report, it is indicated that some features such as PREV.TACHY, TACHYCARDIA, and H.R.DIFF have missing values, with PREV.TACHY being the most affected with a high percentage of 44.02%. In cases where the percentage of missing values is high, it is not recommended to completely delete the feature. Instead, a missing indicator feature can be created to avoid losing important information. 

However, in this case, it is recommended to remove the feature as it contains a significant number of missing values. On the other hand, the TACHYCARDIA feature is the dataset's goal feature, and even a small percentage of missing values (2.01%) can be a concern, and thus the corresponding instances should be eliminated. For the H.R. DIFF feature, the percentage of missing values (13.03%) is not high enough to warrant the removal of the entire case, but the missing values can be replaced using mean imputation for continuous features.

9b) Irregular Cardinality

The dataset contains numeric features with cardinalities much smaller than the number of rows. Continuous values typically have specified ranges. For example, the age bar plot shows only seven distinct ages, indicating a low cardinality for this numeric feature. It is advisable to treat age as a categorical variable and use age ranges such as 10-19 and 20-29 bins, rather than treating age as a numeric feature directly. In the case of the gender feature, the gender bar plot shows four categories, 'f', 'female', 'm', and 'male', giving the gender feature a cardinality of four. This is a data consistency issue. To address this issue, 'f' should be labeled as 'female' and 'm' should be labeled as 'male', which is a common issue with columns such as gender.

9c) Outliers

Outliers are values that are significantly different from the central tendency of a feature, and there are two types of outliers that can occur in an ABT: invalid and valid. Invalid outliers are values that were unintentionally included in a sample and are typically referred to as data noise, while valid outliers are legitimate feature values that are simply extremely different from the other values. To identify outliers in a dataset, the data quality report can be used in two ways. The first is to examine the minimum and maximum values of each feature and determine if they are reasonable based on domain knowledge. The second method for detecting outliers is to look at the difference between the median, lowest, maximum, 1st quartile, and 3rd quartile values.

In the given dataset, HEIGHT, BMI, and SYS.BP have outlier issues. For example, the histogram of HEIGHT feature values has a strong leftward skew. There is a significant gap between the 1st quartile and median values, as well as between the minimum and 1st quartile values. Upon examining the data, we can see that there are a few height values that are significantly lower than the majority of the other values. Several machine learning approaches do not perform well in the presence of outliers.

BMI values should range from about 16 to 60, but there are some very large values in the dataset, with a maximum value of 596,495.39. As BMI is calculated using WEIGHT and HEIGHT, incorrect HEIGHT values are likely to result in incorrect BMI values. The inaccurate HEIGHT values should be corrected and BMI values should be recalculated to eliminate these outliers.

A SYS.BP value of 1,144 is unreasonable since a systolic blood pressure of 140 mm Hg or higher is considered high. This value should be corrected.
