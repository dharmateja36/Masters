5a) A range normalization that generates data in the range (0, 1)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame([42,47,59,27,84,49,72,43,73,59,58,82,50,79,89,75,70,59,67,35],columns=['score'])
h=1
l=0
norm_values = []

for i in df['score']:
    norm_score = ((i - min(df['score']))/((max(df['score'])-min(df['score'])))*(h - l)) + l
    norm_values.append(norm_score)

norm_df=pd.DataFrame(norm_values,columns=['SCORE'])
norm_df
```

9c) Outliers

Ans: Outliers are data values that significantly deviate from the typical central tendency of a feature, and they can be classified into two types: invalid and valid outliers. Invalid outliers refer to data noise that was unintentionally included in a sample, while valid outliers represent legitimate feature values that are simply much different from the other values. There are two methods for detecting outliers in an ABT. The first is examining the minimum and maximum values of each feature based on domain knowledge, while the second is by analyzing the difference between the median, lowest, maximum, 1st quartile, and 3rd quartile values.

In the provided dataset, the HEIGHT, BMI, and SYS.BP features have outlier issues. For instance, the histogram of HEIGHT feature values shows a strong leftward skew, with a significant gap between the 1st quartile and median values, as well as between the minimum and 1st quartile values. Upon inspection, a few height values are significantly lower than the majority of the other values, leading to an outlier issue. Some machine learning methods may perform poorly when there are outliers.

BMI values should range from about 16 to 60, but the dataset contains some extremely large values, with a maximum value of 596,495.39. Since BMI is calculated using WEIGHT and HEIGHT, incorrect HEIGHT values can lead to incorrect BMI values. The incorrect HEIGHT values need to be corrected, and BMI values should be recalculated to eliminate the outliers.

A SYS.BP value of 1,144 is unreasonable since a systolic blood pressure of 140 mm Hg or higher is considered high. This value needs to be corrected.
