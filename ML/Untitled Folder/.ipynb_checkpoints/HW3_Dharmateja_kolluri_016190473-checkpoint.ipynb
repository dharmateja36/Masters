#### a) What target level would a nearest neighbor model using Euclidean distance return for the following email: “machine learning for free”?

```python
s = np.array((0,1,1,0,0,1,1))
id1 = np.array((3,0,0,0,0,0,0))
id2 = np.array((1,2,1,1,1,0,0))
id3 = np.array((0,0,1,1,1,0,0))
id4 = np.array((0,0,1,0,3,1,1))
id5 = np.array((0,1,0,0,0,1,1))
```

```python
def eu_dist(x,y):
    distance = np.sqrt(np.sum((x-y)**2))
    return round(distance,4)
print('Eculidean Distance for id1 ' , eu_dist(id1,s))
print('Eculidean Distance for id2 ' , eu_dist(id2,s))
print('Eculidean Distance for id3 ' , eu_dist(id3,s))
print('Eculidean Distance for id4 ' , eu_dist(id4,s))
print('Eculidean Distance for id5 ' , eu_dist(id5,s))
```

#### e) There are a lot of zero entries in the spam bag-of-words dataset. This is indicative of sparse data and is typical for text analytics. Cosine similarity is often a good choice when dealing with sparse non-binary data. What target level would a 3-NN model using cosine similarity return for the query?

Ans: Consine similarity could be calculated using following formula for the above query:


```python
def cos_sim(x,y):
    similarity = sum(x*y)/((np.sqrt(np.sum((x)**2)))*(np.sqrt(np.sum((y)**2))))
    return round(similarity,4)
s = np.array((0,1,1,0,0,1,1))
id1 = np.array((3,0,0,0,0,0,0))
id2 = np.array((1,2,1,1,1,0,0))
id3 = np.array((0,0,1,1,1,0,0))
id4 = np.array((0,0,1,0,3,1,1))
id5 = np.array((0,1,0,0,0,1,1))
print('Cosine Similarity for id1 ' , cos_sim(id1,s))
print('Cosine Similarity for id2 ' , cos_sim(id2,s))
print('Cosine Similarity for id3 ' , cos_sim(id3,s))
print('Cosine Similarity for id4 ' , cos_sim(id4,s))
print('Cosine Similarity for id5 ' , cos_sim(id5,s))
```

From the above results, we can see that id5, id2 and id4 have the high similarity values as the k value is 3. Since id5 and id4 has SPAM value as False, the result of this model based on cosine similarity would be "False"

#### a) . The company has decided to use a similarity-based model to implement the recommender system. Which of the following three similarity indexes do you think the system should be based on?


Ans: Jaccard similarity is commonly used as a similarity-based model for implementing recommender systems. The Jaccard similarity coefficient is a measure of similarity between two sets of items, which is often used to calculate the similarity between two users or two items in a recommender system.

Russel-Rao and Sokal-Michener are also similarity measures that can be used in recommender systems, but they are less commonly used compared to Jaccard similarity. Russel-Rao similarity is a binary similarity measure that counts the number of items that are common between two sets, while Sokal-Michener similarity is a weighted similarity measure that gives more weight to items that are more rare. However, Jaccard similarity is generally considered to be more effective and accurate for recommender systems.

```python
iris = datasets.load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['target'] = iris.target
data.head()
```

```python
# split the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

# define function to get k nearest neighbors
def get_neighbors(x_train, y_train, x_test_instance, k):
    distances = []
    for i in range(len(x_train)):
        dist = eu_dist(x_train[i], x_test_instance)
        distances.append((x_train[i], y_train[i], dist))
    distances.sort(key=lambda x: x[2])
    neighbors = []
    for i in range(k):
        neighbors.append((distances[i][0], distances[i][1]))
    return neighbors

# make predictions on test set
predictions = []
for i in range(len(x_test)):
    neighbors = get_neighbors(x_train, y_train, x_test[i], k=3)
    output_values = [neighbor[1] for neighbor in neighbors]
    prediction = max(set(output_values), key=output_values.count)
    predictions.append(prediction)

# calculate accuracy
print('The accuracy score of KNN is: ',round(accuracy_score(y_test, predictions),4))
```

#### Weighted KNN function without using KNeighborsClassifier

```python
def predict(x_train, y , x_test, k, w):
    label = []
    for i in x_test:
        dist_p = []
        for j in range(len(x_train)):
            distances = mh_dist(np.array(x_train[j,:]), i)
            dist_p.append(distances)
        dist_p = np.array(dist_p)
        dist = np.argsort(dist_p)[:k]
        labels = y[dist]
        l = mode(labels,keepdims=True) 
        l = l.mode[0]
        label.append(l)
    return label

x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)
predictions = predict(x_train, y_train, x_test, 5, 'distance')
print('The accuracy score of Weighted KNN is: ',round(accuracy_score(y_test, predictions),4))
```

#### KNN Using Sci-kit Learn

```python
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print('The accuracy score of KNN is:', accuracy)
```
