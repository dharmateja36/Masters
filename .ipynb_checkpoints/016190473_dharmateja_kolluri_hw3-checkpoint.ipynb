```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy as sc
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/dharm/Documents/MS/Data Visualization/Assignments/Train.csv')
```

```python
df.head(10)
```

### Histogram with 10 bins for the cost of the Product and its count

```python
plt.figure(figsize = (7,5))
plt.hist(df['Cost_of_the_Product'],bins=10,histtype='bar')
plt.xlabel('Cost of the product')
plt.ylabel('Count')
```

### KDE plot for the cost of product

```python
plt.figure(figsize = (7,5))
sns.kdeplot(df['Cost_of_the_Product'],fill='True')
```

### Stacked KDE plot with cost of the product and product importance

```python
plt.figure(figsize = (9,5))
sns.kdeplot(x=df['Cost_of_the_Product'], hue=df['Product_importance'],multiple='stack',common_norm='True')
```

### Ecdf plot for Discount Offered plotted against Y-axis with proportion as X-axis

```python
plt.figure(figsize = (6,6))
sns.ecdfplot(y= df['Discount_offered'],data=df)
```

### Boxplot for each warehouse block against Cost of the Product

```python
plt.figure(figsize = (10,10))
sns.boxplot(x= df['Warehouse_block'],y=df['Cost_of_the_Product'],order=('A','B','C','D','F'))
```

### Violin plot for the Cost of the Product with Product Importance and Gender of the customer

```python
plt.figure(figsize = (8,5))
sns.violinplot(x='Product_importance',y='Cost_of_the_Product',hue='Gender',data=df)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```
