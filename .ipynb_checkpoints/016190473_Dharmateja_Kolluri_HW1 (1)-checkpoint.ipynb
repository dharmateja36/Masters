```python
df1['Days to Ship'] = (df1['Ship Date'] - df1['Order Date'])/np.timedelta64(1,'D')
df1.sort_values(by=['Days to Ship'],ascending=True).iloc[:10] [['Customer Name','Item ID','Order Date','Ship Date','Ship Mode','Days to Ship']]
```

#### List of values with minimum days taken to ship the products

```python
df1['Order Date Year'] = df1['Order Date'].dt.year 
```

## Using Matplotlib Module

#### 1) Plot against total sales with Category using simple bar graph

```python
plt.xlabel('Total Sales')
plt.title('Category Sales')
df1['Category'].value_counts().plot(kind='barh',figsize=(9,9),color='#a2add0')
```

#### 2) Pie Chart to show the shares of each region for the sales

```python
plt.title('Shares from each region')
df1['Region'].value_counts().plot(kind='pie',figsize=(7,7),autopct='%2.2f%%')
```

```python
fig, axs = plt.subplots(figsize = (15,8))
data_category = df1.groupby(['Order Date Year','Department'])['Profit'].sum()
data_category = data_category.to_frame().reset_index()
sns.lineplot(x='Order Date Year', y='Profit', hue='Department',data= data_category)
```

#### 3) Bar graph to depict number of sales from each state

```python
plt.title('Sales from each state')
plt.xlabel('No. of Sales')
df1['State'].value_counts().plot(kind='barh',figsize=(10,10),color='#25c1bf')
```

#### 4) Graph to depict profit gained from each state

```python
df1[['State','Profit']].groupby('State').sum().sort_values(by=['Profit']).plot(kind='barh',figsize=(10,10),color='#25c1bf')
plt.title('Profits from each state')
plt.xlabel('Profit')
plt.legend(loc='lower right',fontsize='x-large')
```

#### 5) Scatter plot to depict Profit against the Product Base Margin

```python
plt.scatter(df1['Product Base Margin'],df1['Profit'])
plt.title('Plot against Product Base Margin and Profit')
plt.xlabel('Product_Base_Margin')
plt.ylabel('Profit')
plt.show()
```

#### 6) Plot to show the sales for each Region and its underlying States

```python
fig, axs = plt.subplots(figsize = (20,20))
sns.countplot(x="Region", hue ='State', data = df1)
plt.legend(bbox_to_anchor =(0.65, 0.98))
axs.set_title('Region wise Sales', fontsize = 20)
axs.ylabel('Sales')
plt.show()
```

#### 7) Horizontal grouped bar chart with Categories plotted against Sales and Profit for comparison

```python
fig, ax = plt.subplots(figsize = (10,10))
Category_profit = (df1.groupby(['Category'])['Profit'].sum())/100
Category_sales = (df1.groupby(['Category'])['Sales'].sum())/100
Categ = df1['Category'].unique()
bar_width = 0.3
locs = np.arange(len(Category_sales))
ax.barh(locs-bar_width,Category_sales,bar_width,color='#a2add0',label='Sales')
ax.barh(locs,Category_profit,bar_width,color='#fed8b1',label='Profit')
ax.set_yticks(ticks=range(len(Categ)))
ax.set_yticklabels(Categ)
ax.set_xlabel('Total amount/100')
plt.title('Comparison between sales and Profits for each Category')
plt.legend(fontsize= 'large')

#ax[0].bar(color='#25c1bf')
#ax[1].bar(df1['Customer Segment'],df1['Profit'],color='#25c1bf')
plt.show()
```
